```python
import matplotlib.pyplot as plt 
```

```python
X = range(100)
Y = range(100)
plt.plot(X,Y) # plot으로 선그래프 긋기 

#그림 (figure)는 그래프를 작성하는 밑바탕이 되는 툴 
#축 (axes)는 일종의 캔버스로 실제 그래프를 작성하는 공간. 
```

```python
import numpy as np 
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1] # X_1의 코사인 함수 값

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_1] # y_2의 사인 함수 값 

plt.plot(X_1,Y_1) # 코사인 함수 
plt.plot(X_2,Y_2) # 사인 함수 
plt.show()

#plot함수 사용시 한 번에 그래프를 작성가능 
```

```python
fig,ax = plt.subplots() # figure와 axes 객체 할당 

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

ax.plot(X_1,Y_1) #ax객체에 그래프를 작성하는 plot 함수를 사용해서 그래프를 생성 
ax.set(title='cos graph', xlabel='X', ylabel='Y') #그래프 제목과 X,Y축 설정
plt.show()

fig = plt.figure() #fig 반환 
fig.set_size_inches(10,10) # fig 크기 지정 

ax_1=fig.add_subplot(1,2,1) # 첫 번째 그래프 생성 (1행 2열로 figure 객체 공간 생성하고 그 중 1번 열)
ax_2=fig.add_subplot(1,2,2) # 두 번째 그래프 생성 (1행 2열로 figure 객체 공간 생성하고 그 중 2번 열 

ax_1.plot(X_1,Y_1,c="b") # 첫 번째 그래프 설정 (c= 선 색깔)
ax_2.plot(X_2,Y_2,c="g") # 두 번째 그래프 설정
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

x= np.linspace(-1,1,100) # -1부터 100까지 1의 간격으로 
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig,ax = plt.subplots(2,2) #2x2 figure 객체 fig,ax 생성 

ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
plt.show()

```

```python
fig = plt.figure()
fig.set_size_inches(10,10) # figure 크기 설정 
fig.suptitle('The Two Graph') # 메인 figure 제목 
plt.title("Two Graph") # 전체 figure에 대한 공통된 제목 

X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value+100 for value in X_2]

ax1=fig.add_subplot(1,2,1) # 서브 플롯 1 (1행 2열에 1번)
ax2=fig.add_subplot(1,2,2) # 서브 플롯 2

ax1.plot(X_1,Y_1,c="r",ls="dotted") # ax1 선 색깔 , 라인 형태 (dashed, dotted, solid) 
ax1.set_title('first graph') # ax1의 제목 

ax2.plot(X_2,Y_2,color="#000000",linestyle="dashed")
ax2.set_title('second graph')
 
plt.show()
```

```python
plt.plot(X_1,Y_1,c="r",ls="dashed",label="line_1")
plt.plot(X_2,Y_2,c="b",ls="solid",label="line_2")

plt.legend(shadow=True, fancybox=True, loc="upper left")
#shadow = 그림자 효과 , facybox = 주석 모서리 둥글게, , loc = 주석 위치 
plt.title('$y =ax+b$') # $ = 기울임 글씨체 
plt.xlabel('$x_line$')
plt.ylabel('$y_line$')
```

```python
#산점도 
data_1 = np.random.rand(512,2) # 512개의 점을 가진 2차원 배열 
data_2 = np.random.rand(512,2)

plt.scatter(data_1[:,0],data_1[:,1], c="b", marker="x") #data_1의 모든 행에 대해서 0번째 열을 x로, 1번째 열을 y로
plt.scatter(data_2[:,0],data_2[:,1], c="r", marker="o") #data_2의 모든 행에 대해서 0번째 열을 x로, 1번째 열을 y로  
plt.show()
```

```python
N= 50 
X = np.random.rand(N)
Y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15*np.random.rand(N))**2
plt.scatter(X,Y,s=area,c=colors,alpha=0.5) # s는 점의 크기, alpha는 원의 불투명도 
plt.show()
```

```python
#막대 그래프 
data = [[5.,25.,50.,20.],
        [4.,23.,51.,17],
        [6.,22.,52.,19]]

# X 좌표 시작점 
X = np.arange(0,8,2) # 0,2,4,6

# 위치 , 데이터, 그래프 색, 너비 
plt.bar(X+0.00, data[0], color ="b", width= 0.5) # 0.0 , 2.0 , 4.0 , 6.0 // 5,25,50,20 데이터 = 파란색 
plt.bar(X+0.50, data[1], color ="r", width=0.5) # 0.5, 2.5, 4.5, 6.5 // 4,23,51,17 데이터 = 빨간색 
plt.bar(X+1.0, data[2], color="g", width=0.5) # 1 , 3 , 5 , 7  // 6,22,52,19 데이터 = 초록색 

plt.xticks(X+0.05, ("A","B","C","D")) # X축에 표시될 이름과 위치 

plt.show()
```

```python
#누적 막대 그래프 

data = np.array([[5,25,50,20],[4,30,55,25],[2,22,52,19]])
color_list=['b','g','r']
data_label=["A","B","C"]
X=np.arange(data.shape[1])

data = np.array([[5,5,5,10],[2,4,6,8],[6,22,10,19]])

for i in range(3):
    plt.bar(X,data[i], bottom=np.sum(data[:i], axis=0), color=color_list[i], label=data_label[i])
    ## 막대 상에서 아랫부분까지의 값을 더한 다음 그 값을 bottom에 넣는다. 
plt.legend() # 주석 
plt.show()
```

```python
#히스토그램 
N=100
X=np.random.normal(size=N) # normal = 정규 분포 ( 정규 분포에서 random하게 가져옴 - 평균 주변의 값 가져옴 )

plt.hist(X,bins=10) # bins = 막대의 개수 
plt.show()
```
