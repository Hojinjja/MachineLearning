```python
#판다스는 데이터 테이블 형태의 2차원 데이터를 다룬다.
#-판다스는 기본적으로 넘파이를 다루고 wrapper 라이브러리로 사용한다. 
#데이터 프레임 = 데이터 테이블 전체를 다루는 객체 
#시리즈 = 리스트,튜블,넘파이 배열의 1차원 자료구조 - 데이터프레임의 한 행 or 열 
#시리즈 객체 - >피쳐 벡터와 같은 개념으로 사용. 
#인덱스 -> 항상 0부터 시작하고 숫자로만 할당 가능했으나, 시리즈 객체에서는 문자열도 가능. 
#데이터 타입 -> 넘파이의 데이터 타입과 일치함. 
```

```python
import pandas as pd 
import numpy as np 
from pandas import Series, DataFrame 

list_data = [1,2,3,4,5] 
list_name = ["a","b","c","d","e"]
example_obj = Series(data = list_data,index = list_name) #data로는 list_data를 쓰고, index로는 list_name을 사용 
print(example_obj,"\n")
print(example_obj.index,"\n")
print(example_obj.values,"\n")
print(type(example_obj.values))
```

```python
example_obj.name="number" #데이터 프레임 이름 지정 
example_obj.index.name="id" #인덱스 이름 지정 
example_obj
```

```python
dict_data = {"a":1,"b":2,"c":3,"d":4,"e":5} #키 값은 인덱스로 사용
example_obj = Series(dict_data, dtype=np.float32, name="example_data") #value 값은 데이터로 사용 ,datatype, 데이터 열 이름 지정 가능 
example_obj
```

```python
## 그룹별 집계 

import pandas as pd # pandas 모듈 호출
import numpy as np # numpy 모듈 호출

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'Kings', 'Kings', 'Kings',            'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
            'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
            'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data) # 데이터 프레임 형태로
df
```

```python
df.groupby("Rank")["Points"].mean() ## points 기준으로 sum 연산 (sum, count, mean) 가능
```

```python
multi_groupby = df.groupby(["Team", "Year"])["Points"].sum() ##df(데이터 프레임 객체)를 "Team","Year"에 따라 그룹화 "Points"열은 연산의 대상
multi_groupby
```

```python
# 병합
import pandas as pd
#과목 번호, 점수
raw_data = {
    'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
    'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

df_left = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_left
```

```python
#과목 번호, 이름
raw_data = {
    'subject_id': ['4', '5', '6', '7', '8'],
    'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

df_right = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_right
```

```python
#subject_id를 기준으로 내부 조인
#left에는 6번이 없는데, right에는 6번이 있음 -> 내부 조인 이므로 6번은 버려짐
pd.merge(left=df_left, right=df_right, how="inner", on='subject_id')
```

```python
pd.merge(
    df_left, df_right,
    on='subject_id', how='left') # 왼쪽 조인-> how를 right로 바꾸면 오른쪽 조인
```

```python
pd.merge(df_left, df_right, on='subject_id', how='outer') #완전 조인 => outer
```

```python
df_left.index = df_left.subject_id ## df_left와 right의 인덱스를 subject_id로 설정하고, 데이터에서의 subject_id는 삭제
del df_left["subject_id"]
df_right.index = df_right.subject_id
del df_right["subject_id"]

pd.merge(df_left, df_right, on='subject_id', how='inner') # 내부 조인으로 -> 키가 둘 다 존재하는 것만
```
